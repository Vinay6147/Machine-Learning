```python
days = 365

dt=1/365

mu=rets.mean()['GOOG']

sigma=rets.std()['GOOG']
```

```python
def func(dt):
    return np.random.normal(loc=mu*dt,scale=sigma*np.sqrt(dt),size=None)
```

```python
def stock_monte_carlo(start_price,days,mu,sigma):
    
    price=np.zeros(days)
    price[0]=start_price
    
    for i in range(1,days):
        
        price[i]=price[i-1]*(1+(mu*dt)+func(dt))
        
    return price           
```

```python
start_price = 110.1039

for run in range(100):
    
    plt.plot(stock_monte_carlo(start_price,days, mu, sigma))
plt.xlabel('Days')
plt.ylabel('Price')
plt.title('Monte Carlo Analysis for Google')
```

```python
runs=10000

simulations = np.zeros(runs)

np.set_printoptions(threshold=5)

for run in range(runs):
      simulations[run] = stock_monte_carlo(start_price,days,mu,sigma)[days-1]
    
```

```python
q = np.percentile(simulations,1)

plt.hist(simulations,bins=200,color = 'blue')

plt.figtext(0.6,0.8,s='Start price: %f' %start_price)
plt.figtext(0.6,0.7,'Mean final price: %f' %simulations.mean())
plt.figtext(0.6,0.6,'Var(0.99): %f' %(start_price-q,))
plt.figtext(0.15,0.6,'q(0.99): %f' %q)

plt.axvline(x=q,linewidth=4,color='r')

plt.title(u'Final price distribution for Google stock after %s days' %days,weight='bold');
```
